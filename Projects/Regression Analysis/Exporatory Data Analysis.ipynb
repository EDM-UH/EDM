```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
pd.options.display.max_columns = 100

%matplotlib inline
```

```python
file_path = 'data/grades_cleaned_20170520.csv'
grades = pd.read_csv(file_path)
grades.head()
```

```python
grades.describe()
```

# Attendance data looks highly irregular and doesn't make much sense

```python
attendance_mask = grades.columns.str.contains('attend')
attendance = grades.loc[:,attendance_mask]
ax = attendance.mean().plot()
yticks = ax.get_yticks()
ax.set_yticklabels(['{:3.0f}%'.format(tick * 100) for tick in yticks])
ax.set(ylabel = '% attendance', xlabel = 'date')
```

# Blackboard quiz grades appear as expected

```python
quiz_mask = grades.columns.str.contains('quiz') & ~grades.columns.str.contains('class')
quiz = grades.loc[:,quiz_mask]
quiz.hist(figsize = (8,8))
```

# Blackboard quiz participation rose throughout the semester

```python
quiz_attempt = quiz.isnull()
ax = quiz_attempt.mean().plot(ylim = (0,.65))
ticks = ax.get_yticks()
ax.set_yticklabels(['{:3.0f}%'.format(tick*100) for tick in ticks])
ax.set(ylabel = 'quiz participation rate', xlabel = 'quiz')
```

# Homework grades make sense as well. HW's 1 and 3 are based on 'infinite arithmetic' program

```python
homework_mask = grades.columns.str.contains('homework')
homework = grades.loc[:,homework_mask]
ax = homework.mean().plot(ylim = (0, 90), kind = 'area', alpha = 0.5, figsize = (10,3))
ax.set(xlabel = 'assignment', ylabel = 'class avg')
```

# Homework grades are not normally distributed. Students were either able to do it or not, with very few in between.

```python
axes = homework.hist(figsize = (10,5), layout = (2,3))
[ax.set(ylabel = 'number of students',xlabel = 'grade') for ax in axes.reshape(axes.size)]
plt.tight_layout()
```

# The first exam was the most difficult

```python
exam_mask = grades.columns.str.contains('exam')
exams = grades.loc[:, exam_mask]
ax = exams.mean().plot(kind = 'area', ylim = (0,80), alpha = 0.5)
ax.set(ylabel = 'class avg', xlabel = 'grades')
```

# First exam normally distributed, second and third heavily skewed leftwards

```python
axes = exams.hist(layout = (1,3), figsize = (12,3), bins = 10)
[ax.set(xlabel = 'exam grade', ylabel = 'number of students') for ax in axes.reshape(axes.size)]
```
