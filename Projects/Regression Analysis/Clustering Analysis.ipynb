```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
pd.options.display.max_columns = 100

%matplotlib inline
```

```python
file_path = 'data/grades_cleaned_20170520.csv'
grades = pd.read_csv(file_path)
grades.head()
```

# Create dataframe of grade summary statistics

```python
# Create series consisting of each students homework average
homework_avg = grades.loc[:,grades.columns.str.contains('homework')].mean(axis = 1)
homework_std = grades.loc[:,grades.columns.str.contains('homework')].std(axis = 1)

# Create series consisting of each students exam average
exam_avg = grades.loc[:,grades.columns.str.contains('exam')].mean(axis = 1)
exam_std = grades.loc[:,grades.columns.str.contains('exam')].std(axis = 1)

# Create series counting number of blackboard quiz attempts. The exact grade doesn't matter because they were optional.
quiz_count = grades.loc[:,grades.columns.str.contains('quiz') & ~grades.columns.str.contains('class')].count(axis = 1)

summary = pd.DataFrame(np.array([homework_avg, homework_std, exam_avg, exam_std, quiz_count, grades['0510_group'],
                                 grades['0510_extra'], grades.total_weighted]).T,
                       columns = ['homework_avg','homework_std','exam_avg','exam_std','quiz_count','group','extra',
                                  'total_weighted'])
summary.head()
```

### One student turned in 52 extras. This is an outlier that affect the clustering, so we adjust it down.

```python
summary.extra.value_counts()
```

```python
summary['extra'] = summary['extra'].apply(lambda x: x if x < 50 else 18.0)
```

# Correlation analysis on summary statistics
- Homework average (without dropping lowest grade) more correlated with course grade than exam grades despite being weighted less.
- Number of blackboard quiz attempts not correlated with final grade.

```python
sns.heatmap(summary.corr(), annot = True)
```

# Principal component analysis
Here, we project an 8-dimensional vector space onto a 2-dimensional one, preserving 61.72% of the variance of the data. At the very least, there appear to be three clusters.

```python
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler

pca = PCA(n_components=3)
norm = StandardScaler()

summary_normalized = norm.fit_transform(summary)
summary_3d = pca.fit_transform(summary_normalized)

print('{:2.2f}% variance explained by 2 components'.format(pca.explained_variance_ratio_[:2].sum() * 100))

sns.jointplot(summary_3d[:,0], summary_3d[:,1])
```

### Project the data onto a 3-dimensional vector space using PCA

```python
%matplotlib inline

print('{:2.2f}% variance explained by 3 components'.format(pca.explained_variance_ratio_[:3].sum() * 100))

from mpl_toolkits.mplot3d import Axes3D
fig = plt.figure()
ax = Axes3D(fig)
ax.scatter(summary_3d[:,0], summary_3d[:,1], summary_3d[:,2])
ax.view_init(azim = 45, elev = 45)
```

```python
%matplotlib inline

tsne3d = TSNE(n_components=3, early_exaggeration = 10, learning_rate = 100, perplexity = 5)
summary_tsne3d = tsne3d.fit_transform(summary_normalized)

fig = plt.figure()
ax = Axes3D(fig)
ax.scatter(summary_tsne3d[:,0], summary_tsne3d[:,1], summary_tsne3d[:,2])
```

### Let's see what these clusters look when projected onto a 3-dimensional vector space using PCA
- The cluster structures themselves could be stronger, but for the most part this seems to make sense.

```python
%matplotlib inline
from sklearn.cluster import KMeans
from matplotlib import colors

kmeans = KMeans(n_clusters = 6)
kmeans.fit(summary_normalized)
clusters = kmeans.labels_

fig = plt.figure()
ax = Axes3D(fig)
g = ax.scatter(summary_3d[:,0], summary_3d[:,1], summary_3d[:,2], c = clusters, cmap = 'plasma')
plt.colorbar(g, ax = ax)
ax.view_init(azim = 45, elev = 45)
```

# What do the clusters mean?
From the plots below, we can make some characterizations about each cluster.
- **Cluster 1 and 2**: These are the "overachiever" A and B students. Together, they would comprise the second largest cluster and have the highest course grades, no doubt buoyed by their high homework and exam averages. With the exception of some outliers, they all have 100's on their group assignments and notably have low homework standard deviations. Most in this cluster keep up with optional blackboard quizzes, though a few do not, probably because with decent grades, there is no to take optional quizzes. The distinguishing feature between Cluster 1 and 2 students is that Cluster 2 students turned in extra assignments.
- **Cluster 6**: There are the all F "hopeless" students. They are characterized by the lowest course grades, homework averages, exam averages, and group grades. Cluster 6 students are the only ones to have very high exam standard deviations, probably because they missed at least one exam.
- **Cluster 5**: This cluster consists of "mostly hopeless" F students. It is only slightly better than Cluster 6 across all metrics. Essentially, Cluster 0 students are Cluster 6 students who at least took all exams.
- **Cluster 4**: This cluster consists of the "typical" F, D, and C students. In terms of course grade, homework average, exam average, and group grade, Cluster 4 is near the middle of all six clusters. However, Cluster 4 students are the ones besides the "overachievers" in Cluster 2 who participated in the blackboard quizzes. Clearly, these are students who are struggling but at least are trying to do something about it.
- **Cluster 3**: This cluster is interesting and consists of "typical" and "above average" F, D, C, B  students. Cluster 3 is like a better version of Cluster 4 except that no one in Cluster 1 took many blackboard quizzes. The more successful students of this cluster may simply be content with their grades, while the less successful, while possibly not happy with their performance, do not feel inclined to put in extra work.
- **Bottom line**: Focus resources on underperforming Cluster 4 students.
