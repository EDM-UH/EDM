# Section 5: Seaborn

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns

%matplotlib inline
```

## Univariate distributions

```python
tips = sns.load_dataset('tips')
tips.head()
```

### histograms with distplot()

```python
ax = sns.distplot(tips.tip.dropna(), kde = True, rug = True)
ax.set(title = 'Distribution of Tips')
```

### histograms with countplot() for categorical features

```python
sns.countplot(data = tips, x = 'day', hue = 'sex')
```

### boxplot()

```python
sns.boxplot(data = tips, x = 'day', y = 'tip', hue = 'sex')
```

### violinplot()

```python
sns.violinplot(data = tips, x = 'day', y = 'tip', hue = 'sex', split = True)
```

### stripplot()

```python
sns.stripplot(data = tips, x = 'day', y = 'tip', jitter = True)
```

## Bivariate distributions

### jointplot(kind = 'scatter')

```python
sns.jointplot(data = tips, x = 'total_bill', y = 'tip', kind = 'scatter')
```

### jointplot(kind = 'hex')

```python
sns.jointplot(data = tips, x = 'total_bill', y = 'tip', kind = 'hex')
```

### jointplot(kind = 'kde')

```python
sns.jointplot(data = tips, x = 'total_bill', y = 'tip', kind = 'kde')
```

## Pairwise relationships

```python
iris = sns.load_dataset('iris')
iris.head()
```

### correlation plot

```python
print(iris.corr(),'\n')
sns.heatmap(iris.corr(), annot = True, cmap = 'RdBu')
```

### pairplot()

```python
sns.pairplot(data = iris, hue = 'species', diag_kind = 'kde')
```

### a pairplot is a special type of PairGrid object

```python
ax = sns.PairGrid(data = iris)
ax.map_diag(sns.distplot)
ax.map_offdiag(sns.kdeplot)
```

## Summarize with barplot()

```python
sns.barplot(data = tips, x = 'sex', y = 'tip', hue = 'day')
```

### can use other aggregation functions like sum, median, max, and min

```python
sns.barplot(data = tips, x = 'sex', y = 'tip', hue = 'day', estimator = np.std)
```

### y = numeric binary feature results in proportions displayed on y-axis

```python
sns.barplot(x = tips.day, y = tips.sex == 'Male', hue = tips.smoker)
```

## pointplot() shows trends in numerical values from one category to another

```python
sns.pointplot(data = tips, x = 'day', y = 'total_bill', hue = 'sex')
```

## factorplot() allows for even greater division of categorical variables

```python
tips.head(0)
```

```python
sns.factorplot(data = tips, x = 'sex', y = 'total_bill', col = 'day', row = 'time', hue = 'smoker', kind = 'bar')
```
