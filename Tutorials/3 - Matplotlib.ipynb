# Section 3: Matplotlib

```python
import matplotlib.pyplot as plt
import numpy as np

%matplotlib inline
```

## Basic line plot

```python
x = np.arange(0, 7, 0.01)
y = [np.sin(i) for i in x]
plt.plot(x,y)
plt.plot(x,[np.cos(i) for i in x])
plt.grid(True)
plt.legend(['$sin(x)$','$cos(x)$'], loc = 0)
```

## Scatter plot

```python
x = np.linspace(0,6.28,50)
y = [np.cos(i) for i in x]
plt.scatter(x, y, c = y, s = x * 10)
plt.title('this is a scatter plot')
plt.xlabel('notice how size increases with x-value')
plt.ylabel('notice how color changes with y-value')
plt.xlim(0,6.28)
plt.grid(True)
```

## Histogram

```python
things = np.random.randn(10000) * 100 // 1
plt.hist(things, color = 'salmon', alpha = 0.5, bins = 30)
plt.grid(True)
plt.show()
```

## Bar plot

```python
x = [1,2,3,4,5]
y = [20,3,15,8,12]
plt.bar(x,y, align = 'center', color = 'goldenrod')
```

```python
x = [1,2,3,4,5]
y = [20,3,15,8,12]
plt.barh(x,y, align = 'center', color = 'chartreuse')
```

## Subplots!!!!!!!!!!!!

```python
fig, axes = plt.subplots(3,2, figsize = (10,8))
axes[0][0].scatter(np.linspace(0,10,100), np.cos(np.linspace(0,10,100)), color = 'skyblue')
axes[0][0].set(title = 'scatter plot', xlabel = 'x', ylabel = 'y')
axes[0][1].plot(np.arange(100), np.random.randn(100), color = 'darkred')
axes[0][1].set(title = 'earthquake')
axes[1][1].hist(np.random.randn(100), bins = 20, color = 'forestgreen')
axes[1][1].set(title = 'you get the idea')
axes[2][0].contourf(np.arange(10),np.arange(10),np.random.randn(10,10), cmap = 'plasma', levels = np.linspace(-3,3,50))
plt.tight_layout()
plt.show()
```
